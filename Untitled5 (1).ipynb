```python
x=np.random.normal(loc=60,scale=5,size=1000)#x
y=np.random.normal(loc=60,scale=5,size=1000)#y
plt.hist(x,bins=20)
plt.show()
plt.plot(x)
plt.show()
sns.distplot(x,hist=False,label='normal 5')#x
sns.distplot(y,hist=False,label='normal 2')#y
plt.legend()
plt.show()
```

```python
z=np.random.poisson(lam=60,size=1000)
plt.hist(z,bins=20)
plt.show()
plt.plot(z)
plt.show()
sns.distplot(z,hist=False,label='bino')#y
plt.legend()
plt.show()
```

```python
x=np.random.normal(loc=60,scale=5,size=1000)#x
y=np.random.binomial(n=100,p=0.6,size=1000)
z=np.random.poisson(lam=60,size=1000)
sns.distplot(x,hist=False,label='normal')#y
sns.distplot(y,hist=False,label='bino')#y
sns.distplot(z,hist=False,label='poisson')
plt.legend()
plt.show()
```

```python
i=np.random.uniform(low=55,high=65,size=1000)
plt.hist(i,bins=20)
plt.show()
plt.plot(i)
plt.show()
sns.distplot(i,hist=False,label='uniform')#y
plt.legend()
plt.show()
```

```python
x=np.random.normal(loc=60,scale=5,size=1000)#x
y=np.random.binomial(n=100,p=0.6,size=1000)
z=np.random.poisson(lam=60,size=1000)
i=np.random.uniform(low=45,high=75,size=1000)
sns.distplot(x,hist=False,label='normal')#y
sns.distplot(y,hist=False,label='bino')#y
sns.distplot(z,hist=False,label='poisson')
sns.distplot(i,hist=False,label='uniform')
plt.legend()
plt.show()
```
